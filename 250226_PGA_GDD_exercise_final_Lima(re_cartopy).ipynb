# **Growing Degree Days (Exercise)**
Growing Degree Days (GDD) is a basic thermal-based temperature model used in agriculture and biology to predict crop growth and development or understand specific biological events. It measures thermal accumulation above a certain threshold over a given period, making it a valuable tool for tracking the thermal energy required for plant growth.

The growing degree for a given temperature ($T$) represents the difference between that temperature and the base temperature ($T_{base}$) indicating the extent to which a crop responds and grows. However, growth does not increase beyond the optimal temperature ($T_{opt}$) and never falls below zero to prevent negative values. This temperature difference is calculated on a daily basis (**growing degree days**), and is expressed in units of $K d^{-1}$ (or degrees Celsius).

The accumulated value of Growing Degree Days (GDD) over time is referred to as **cumulative growing degree days** (CGDD) or **simply GDD accumulation**. It is commonly used as a benchmark for reaching key growth or developmental stages, such as flowering(or heading) or maturity.



1.   Calcuation GDD & CGDD  
GDD is calculated by accumulating the excess temperature when the average temperature during the day is above the base temperature of the crop. The base temperature varies depending on the crop, refer to the table 1 in this exercise. GDD is calculated using the following formula:
$$
GD(T)= \frac{(T_{max} + T_{min})}{2} - T_{base}
$$

\begin{align}
\mathrm{GDD}(T) &= \frac{\mathrm{GD}(T)}{1\ d} \\
\mathrm{cGDD} &= \sum_i^n \mathrm{GDD}(T_i) \\
\end{align}  



2.   Application GDD   


*   **Crop Growth Prediction:** Estimates when a specific crop will reach key growth stages such as germination, flowering, and maturity for harvest.
*   **Pest Management:** Predicts the emergence and activity of pests and pathogens to determine the best time for control.
*   **Variety Selection:** Helps determine suitable crops for a given region based on cumulative GDD.



3.   Limitations about GDD

*   **Extreme Temperatures:** High temperatures above the base threshold can slow down crop growth, but GDD does not account for this.

*   **Ignoring Environment Factors:** GDD does not consider other environmental factors such as humidity, sunlight, or soil conditions.

**Despite these limitations, GDD remains a widely used, simple, and practical tool for growth and development prediction.**

```python
## define base temperature
T_base = 10
```

```python
## define the parameter to save filtered and merged data
df_final_sheets = {}
```


* As mentioned, GDD is calculated from the daily maximum and minimum air temperatures.  
* When the daily GDD values accumulate over time, this becomes **the cumulative GDD**.  
* The reason cumulative GDD is needed is that, as explained ago, it is essential for understanding growth, development, and maturation.  
* **the Cumulative GDD** provides the information required for these processes.

* So, with calculating the daily GDD, we will also code the accumulation of GDD, i.e., **the cumulative GDD** for each day, at the same time.
* To calculate GDD for all sheets in the Excel file, we will use a **for loop**.

```python
for sheet_name, df in df_asos.items():
    ## GDD equation
    df['Base10C_GDD'] = ((df['TMAX'] + df['TMIN']) / 2 - T_base).clip(lower=0)

    ## calculate the cumulative GDD (at stations)
    df['Accumulated_GDD'] = df.groupby('STNID')['Base10C_GDD'].cumsum()

    ## select columns to save (include GDD with 'time')
    df_final_sheets[sheet_name] = df[['STNID', 'Latitude', 'Longitude', 'TIME', 'TMAX', 'TMIN', 'Base10C_GDD', 'Accumulated_GDD']]
```

* In this exercise, we will calculate the cumulative GDD for two stages: **the reproductive(flowering, anthesis) stage** and **the maturity stage** which is suitable for harvest.  
* We will also estimate the dates when the required cumulative GDD for these two stages is reached.


* For Maize, we will set the required cumulative GDD for the **reproductive (flowering, anthesis)** stage to 800, which is the midpoint between 700 and 900 degrees.  
* For the full **maturity** stage, we will set the required cumulative GDD to 1,400, which falls between 1,300 and 1,500 degrees.


* In the code above, the data is currently stored in memory in **df_final_sheets** and has also been saved in an Excel file.  
* You can either retrieve the required cumulative GDD from the saved Excel file or from **df_final_sheets**.  
* Since, the data is still in memory, we will read it directly from **df_final_sheets** for this exercise.

```python
## define yearly GDD_800, GDD_1400
gdd_800_by_year = {}
gdd_1400_by_year = {}
```

```python
## define year
for year in range(1991, 2021):
    sheet_name = str(year)

    ## read sheet
    try:
        df = df_final_sheets[sheet_name]

        ## sort data as STNID
        for stnid, group in df.groupby('STNID'):

            ## extract the time that GDD reachs 800 and 1400
            time_800 = group[group['Accumulated_GDD'] >= 800].iloc[0] if not group[group['Accumulated_GDD'] >= 800].empty else None
            time_1400 = group[group['Accumulated_GDD'] >= 1400].iloc[0] if not group[group['Accumulated_GDD'] >= 1400].empty else None

            ## if there is no result (not add)
            if time_800 is not None:
                gdd_800_by_year.setdefault(year, []).append([stnid, time_800['Latitude'], time_800['Longitude'], time_800['TIME'], time_800['Accumulated_GDD']])
            if time_1400 is not None:
                gdd_1400_by_year.setdefault(year, []).append([stnid, time_1400['Latitude'], time_1400['Longitude'], time_1400['TIME'], time_1400['Accumulated_GDD']])

    except ValueError:
        ## if there is no sheet for that year, an exception.
        print(f"{sheet_name} Cant find sheet.")
```


* Also, need **dictionary** to save dates converted to **day of year**.

```python
## set two dictionaries: GDD 800 & GDD 1400 to save doy (day of year) that reached GDD 800 and 1400.
gdd_800_dates = {}
gdd_1400_dates = {}

## make color list (Specify different colors for STNID)
colors = plt.cm.get_cmap('tab10', 20)  ## maximum 20 colors
```

```python
## data by year
for year in range(1991, 2021):
    sheet_name_800 = f'{year}_GDD_800'  ## example: 1991_GDD_800
    sheet_name_1400 = f'{year}_GDD_1400'  ## example: 1991_GDD_1400

    try:
        ## read GDD_800
        day_800 = pd.read_excel(gdd_Date_output, sheet_name=sheet_name_800)
        ## read GDD_1400
        day_1400 = pd.read_excel(gdd_Date_output, sheet_name=sheet_name_1400)

        ## extract dates that reached GDD_800 and GDD_1400 by STNID from each sheet
        for stnid, group_800 in day_800.groupby('STNID'):
            time_800 = group_800[group_800['Accumulated_GDD'] >= 800].iloc[0] if not group_800[group_800['Accumulated_GDD'] >= 800].empty else None
            time_1400 = day_1400[day_1400['STNID'] == stnid]
            time_1400 = time_1400[time_1400['Accumulated_GDD'] >= 1400].iloc[0] if not time_1400[time_1400['Accumulated_GDD'] >= 1400].empty else None

            ## doy: reached GDD 800
            if time_800 is not None:
                day_of_year_800 = time_800['TIME'].dayofyear
                gdd_800_dates.setdefault(stnid, []).append((day_of_year_800, time_800['TIME']))

            ## DOY: reached GDD 1400
            if time_1400 is not None:
                day_of_year_1400 = time_1400['TIME'].dayofyear
                gdd_1400_dates.setdefault(stnid, []).append((day_of_year_1400, time_1400['TIME']))

    except ValueError:
        print(f"{sheet_name_800} or {sheet_name_1400} cannot find the sheet.")  ## as exception
```
